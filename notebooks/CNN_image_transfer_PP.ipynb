```python
import pandas as pd
from sklearn.utils import shuffle
import os
from sklearn.model_selection import train_test_split

```

```python
import shutil
```

```python
#upload labels
df_labels = pd.read_csv('/Users/ppellegrini/code/PasPelle/ML_vs_Cancer/raw_data/train_labels.csv')
df_labels
```

```python
#size of each class

sample_n = 10000

#create stable sets of healthy vs metastasis and shuffle them into one dataframe

df_0= pd.DataFrame(df_labels[df_labels['label']==0].sample(sample_n,random_state=42))

df_1= pd.DataFrame(df_labels[df_labels['label']==1].sample(sample_n,random_state=42))

```

```python
#merge the 2 dataframes in 1

df = pd.concat([df_0, df_1], axis=0)

df = shuffle(df)

df
```

```python
# save to csv file
df.to_csv('10k_subset_labels.csv', index=False)
```

```python
# # Create a new directory so that we will be using the ImageDataGenerator
base_dir='/Users/ppellegrini/code/PasPelle/ML_vs_Cancer/raw_data/'
# os.mkdir(base_dir)

# create a path to 'base_dir' to which we will join the names of the new folders
# train_dir
train_dir = os.path.join(base_dir, 'train_subset')

# val_dir
val_dir = os.path.join(base_dir, 'val_subset')

# Check and create train directory if it doesn't exist
if not os.path.exists(train_dir):
    os.mkdir(train_dir)
    print(f"Train directory '{train_dir}' created.")
else:
    print(f"Train directory '{train_dir}' already exists.")

# Check and create validation directory if it doesn't exist
if not os.path.exists(val_dir):
    os.mkdir(val_dir)
    print(f"Validation directory '{val_dir}' created.")
else:
    print(f"Validation directory '{val_dir}' already exists.")

```

```python
# [CREATE FOLDERS INSIDE THE TRAIN AND VALIDATION FOLDERS]
# Inside each folder we create seperate folders for each class

# create new folders inside train_dir
no_tumor_tissue = os.path.join(train_dir, 'a_no_tumor_tissue')

has_tumor_tissue = os.path.join(train_dir, 'b_has_tumor_tissue')

# Check and create directory if it doesn't exist
if not os.path.exists(no_tumor_tissue):
    os.mkdir(no_tumor_tissue)
    print(f"Train directory '{no_tumor_tissue}' created.")
else:
    print(f"Train directory '{no_tumor_tissue}' already exists.")

# Check and create directory if it doesn't exist
if not os.path.exists(has_tumor_tissue):
    os.mkdir(has_tumor_tissue)
    print(f"Validation directory '{has_tumor_tissue}' created.")
else:
    print(f"Validation directory '{has_tumor_tissue}' already exists.")



# create new folders inside val_dir
val_no_tumor_tissue = os.path.join(val_dir, 'a_no_tumor_tissue')
val_has_tumor_tissue = os.path.join(val_dir, 'b_has_tumor_tissue')


# Check and create directory if it doesn't exist
if not os.path.exists(val_no_tumor_tissue):
    os.mkdir(val_no_tumor_tissue)
    print(f"Validation directory '{val_no_tumor_tissue}' created.")
else:
    print(f"Validation directory '{val_no_tumor_tissue}' already exists.")

# Check and create directory if it doesn't exist
if not os.path.exists(val_has_tumor_tissue):
    os.mkdir(val_has_tumor_tissue)
    print(f"Validation directory '{val_has_tumor_tissue}' created.")
else:
    print(f"Validation directory '{val_has_tumor_tissue}' already exists.")



```

```python

# #train test split the shuffled dataset
# #tratify=y creates a balanced validation set.

y = df['label']

df_train, df_val = train_test_split(df, test_size=0.10, random_state=101, stratify=y)
```

```python
# Set the id as the index in df_data
df.set_index('id', inplace=True)
```

```python
# Get a list of train and val images
train_list = list(df_train['id'])
val_list = list(df_val['id'])



# Transfer the train images

for image in train_list:
    
    # the id in the csv file does not have the .tif extension therefore we add it here
    fname = image + '.tif'
    # get the label for a certain image
    target = df.loc[image,'label']
    
    # these must match the folder names
    if target == 0:
        label = 'a_no_tumor_tissue'
    if target == 1:
        label = 'b_has_tumor_tissue'
    
    # source path to image
    src = os.path.join('/Users/ppellegrini/code/PasPelle/ML_vs_Cancer/raw_data/train', fname)
    # destination path to image
    dst = os.path.join(train_dir, label, fname)
    # copy the image from the source to the destination
    shutil.copyfile(src, dst)
```

```python
# Transfer the val images

for image in val_list:
    
    # the id in the csv file does not have the .tif extension therefore we add it here
    fname = image + '.tif'
    # get the label for a certain image
    target = df.loc[image,'label']
    
    # these must match the folder names
    if target == 0:
        label = 'a_no_tumor_tissue'
    if target == 1:
        label = 'b_has_tumor_tissue'
    

    # source path to image
    src = os.path.join('/Users/ppellegrini/code/PasPelle/ML_vs_Cancer/raw_data/train', fname)
    # destination path to image
    dst = os.path.join(val_dir, label, fname)
    # copy the image from the source to the destination
    shutil.copyfile(src, dst)
```
